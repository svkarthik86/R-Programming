# Split-apply-combine data analysis and the summarize() function

Split-apply-combine data analysis and the summarize() function
Many data analysis tasks can be approached using the “split-apply-combine” paradigm: split the data into groups, apply some analysis to each group, and then combine the results. dplyr makes this very easy through the use of the group_by() function, which splits the data into groups. When the data is grouped in this way summarize() can be used to collapse each group into a single-row summary. summarize() does this by applying an aggregating or summary function to each group. For example, if we wanted to group by citrate-using mutant status and find the number of rows of data for each status, we would do:

```r
mtcars%>%head(3)
```

```r
unique(mtcars$cyl)
```

```r
mtcars%>%group_by(cyl) %>% summarize(Min_value=min(mpg))
```

```r
url2="https://raw.githubusercontent.com/svkarthik86/Advanced-python/main/subset-covid-data.csv"
covid_data<-read_csv(url2)
head(covid_data,3)
```

```r
covid_data%>%group_by(continent) %>% summarize(avg.cases=mean(cases),avg.deaths=mean(deaths))
```

```r
groups_data<-covid_data%>%group_by(continent)
```

```r
groups_data%>%summarize(cases=mean(cases),deaths=mean(deaths))
```

```r
ungroup(groups_data)%>%summarize(cases=mean(cases),deaths=mean(deaths))
```

# Relational Data Operations

```r
merge() Function - Merge Two Data Frames
```

```r
merge()
```

```r
months <- c("jan","feb","jan","mar")
year <- c("2018","2018","2019","2019")
temp <- c(4.64,19.68,2.56,36.74)
delhi_temp <- data.frame(months,year,temp)
delhi_temp
```

```r
months2 <- c("apr", "mar", "feb", "jun")
year2 <- c("2018","2018","2019","2019")
temp2 <- c(38.75,37.68,28.56,41.74)
delhi_temp2 <- data.frame(months2,year2,temp2)
colnames(delhi_temp2) <- c("months","year","temp")
delhi_temp2
```

```r
merge_delhi_temp <- merge(delhi_temp,delhi_temp2,by="months")#inner join
merge_delhi_temp
```

```r
delhi_temp%>%inner_join(.,delhi_temp2,by="months")  #dplyr package function
```

```r
merge_delhi_temp <- merge(delhi_temp,delhi_temp2,by="months",all = TRUE)#outer join # all = TRUE
merge_delhi_temp
```

```r
delhi_temp%>%full_join(.,delhi_temp2,by="months")  #dplyr package function
```

```r
merge_delhi_temp <- merge(delhi_temp,delhi_temp2,by="months",all.x=TRUE)#left_outer join
merge_delhi_temp
```

```r
delhi_temp%>%left_join(.,delhi_temp2,by="months")  #dplyr package function
```

```r
merge_delhi_temp <- merge(delhi_temp,delhi_temp2,by="months",all.y=T)#Right_outer join
merge_delhi_temp
```

```r
delhi_temp%>%right_join(.,delhi_temp2,by="months")  #dplyr package function
```

```r
merge_delhi_temp <- merge(delhi_temp,delhi_temp2,by=NULL)#cross join
merge_delhi_temp
```

```r
delhi_temp%>%full_join(.,delhi_temp2,by=NULL)  #dplyr package functionby = character()
```

```r
#Join opeartion Using dplyr:: library
```

```r
dplyr::left_join(delhi_temp,delhi_temp2,by = 'year')
```

```r
dplyr::right_join(delhi_temp,delhi_temp2,by = 'year')
```

```r
dplyr::full_join(delhi_temp,delhi_temp2,by = 'months')
```

```r
dplyr::inner_join(delhi_temp,delhi_temp2,by = 'months')
```
